## Running Commands

```python
commands = []

p0 = GUI.get_position()
for i in range(100):
    
    p1 = p0 + np.array([0,0,50])*np.sin(np.pi*i/10)
    p1 = p1 + np.array([30,0,0])*np.sin(np.pi*i/10)
    p1 = p1 + np.array([0,30,0])*np.sin(np.pi*i/10)
    
    command = p1
    commands.append(command)

GUI.run_commands( commands )
```

```python
import numpy as np
import cv2

def record_video(cam_id=0, record=True ):
    
    cap = cv2.VideoCapture(cam_id)

    fourcc = cv2.VideoWriter_fourcc(*'XVID')
    
    if record:
        fn_out = "Camera_Out\\" 
        #out = cv2.VideoWriter(fn_out,fourcc, 20.0, (640,480))
    else: out=None

    while(cap.isOpened()):
        ret, frame = cap.read()
        if ret==True:
            frame = cv2.flip(frame,0)

            cv2.imshow('frame',frame)
            if cv2.waitKey(1) & 0xFF == ord('q'):
                break
        else:
            break

    # Release everything if job is finished
    if out: out.release()
        
    cap.release()
    cv2.destroyAllWindows()
    
```
